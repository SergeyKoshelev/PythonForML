```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

stats.rv_continuous #непрерывные
stats.rv_discrete  #дискретные
```

```python
#дискретные
rv_bernoulli = stats.bernoulli(p=0.3)
rv_bernoulli.rvs(14)
```

```python
rv_binom = stats.binom(100, p = 0.9)
rv_binom.rvs(9)
```

```python
plt.hist(rv_binom.rvs(80), bins = 10)
```

```python
#непрерывные
a = 5
b = 10

rv_uniform = stats.uniform (a, b - a)
rv_uniform.cdf(11) #функция распределения
```

```python
rv_uniform.pdf(7) #плотность вероятности
```

```python
X = np.linspace(a-2, b+2, 100)
#cdf = rv_uniform.cdf(X)
pdf = rv_uniform.pdf(X)
#plt.plot(X, cdf)
plt.plot(X, pdf)
#plt.ylabel('F(x)')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.ylim([0, 1.5])
plt.title('Title')
```

```python
mu = 2
sigma = 0.5
rv_norm = stats.norm(loc = mu, scale = sigma)
rv_norm.rvs(17)
```

```python
x = np.linspace(0, 4, 100)
cdf = rv_norm.cdf(x)
plt.plot(x, cdf)
```

```python
pdf = rv_norm.pdf(x)
plt.plot(x, pdf)
```

```python
mu = 10
sigmas = [1, 2, 4, 6]
for sigma in sigmas : 
    rv_norm = stats.norm(loc = mu, scale = sigma)
    x = np.linspace(0, 20, 200)
    cdf = rv_norm.pdf(x)
    plt.plot(x, cdf, label = sigma)
plt.legend()
```

```python
mean, var, skew = rv_norm.stats(moments='mvs')
print(mean, var, skew)
```
