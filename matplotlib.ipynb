```python
import pandas as pd
from numpy.random import exponential

df = pd.DataFrame({'x' : range(20), 'y' : exponential(10,20)})
df.y.hist()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

data1 = exponential(5, 20)
data2 = exponential(6, 20)
plt.plot(data1, label = "First player")
plt.scatter(range(len(data2)), data2, color = 'purple', label = "Second player")
plt.title('Results', fontdict={'fontsize' : 15})
plt.xlabel('№ attempt')
plt.ylabel('Result')
plt.legend()
plt.xticks(range(0, 20, 4))
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (10, 5))
for row, row_axes in enumerate(axes) :
    for column, ax in enumerate(row_axes):
        ax.plot(exponential(column, 20))
        ax.set_title("Холст колонка {} строка {}".format(column + 1, row + 1))
        ax.set_xlabel("Attempt")
        ax.set_ylabel('Result')
fig.tight_layout()
```

```python
fig.savefig("all_results.pdf")
```
